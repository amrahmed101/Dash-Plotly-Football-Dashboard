```python
import dash
from dash import dcc
from dash import html
from jupyter_dash import JupyterDash
import plotly.express as px
import pandas as pd
from dash.dependencies import Input, Output
```

<b> Plotting the Graphs </b>

```python

df["Goals/Min"]=df["Goals"]/df["Minutes Played"]
df_effective=df.sort_values(by="Goals/Min",ascending=False)[:10]
drop_down1_options = ['Rating', 'Shots On Target', 'Total Shots', 'Goals', 'Assists', 'Key Passes', 
                      'Dribbles', 'Penalty Scored', 'Penalty Missed']
app=JupyterDash(external_stylesheets=['https://codepen.io/chriddyp/pen/bWLwgP.css'])


app.layout=html.Div(children=[dcc.Markdown("##### Top 5 Football Leagues 2020/21 Top Scorers Statistics",style={"text-align": "center"}),
                     
                     dcc.Graph(id='graph-with-dropdown',style={"border": "1px solid #A1CFEE",
                                                               "border-radius": "5px",'width':'650px',
                                                               "display":"inline-block",
                                                               "margin-right": "2px",
                                                               "margin-left": "10px"})
                     
                     dcc.Graph(id='graph-with-dropdown2',style={"border": "1px solid #A1CFEE","border-radius": "5px",
                                                                'width':'650px',"display":"inline-block",
                                                                "margin-left": "2px","margin-right": "10px"})
                     
                     dcc.Dropdown(id="drop-down",style={'width':'300px',"display":"inline-block",
                                                        "margin-left": "50px"},
                                 options=[{'label': str(name), 'value': str(name)} for name in drop_down1_options],
                                 value='Goals'),
                     
                     dcc.Dropdown(id="drop-down2",style={'width':'325px',"display":"inline-block",
                                                         "margin-left": "200px", "margin-right": "auto"},
                                 options=[{'label':'Team', 'value':'Team Name'},
                                          {'label':'League', 'value':'League Name'}],
                                 value='Team Name'),
                     
                     dcc.Dropdown(id="drop-down3",style={'width':'325px',"display":"inline-block",
                                                         "margin-left": "50px", "margin-right": "auto"},
                                 options=[{'label': str(name), 'value': str(name)} for name in drop_down1_options],
                                 value='Rating'),
                     
                    dcc.Markdown("##### Compare Between Two Player Stats",style={"text-align": "center"}),
                     
                    dcc.Graph(id='graph-with-dropdown3',style={"border": "1px solid #A1CFEE",
                                                               "border-radius": "5px",'width':'650px',
                                                               "display":"inline-block",
                                                               "margin-left": "10px"}),
                     
                    dcc.Graph(id='graph-with-dropdown4',style={"border": "1px solid #A1CFEE",
                                                               "border-radius": "5px",
                                                               'width':'650px',"display":"inline-block",
                                                               "margin-right": "10px"}),
                     
                    dcc.Dropdown(id="drop-down4",style={'width':'300px',"display":"inline-block", "margin-left": "70px", "margin-right": "470px"},
                                options=[{'label':str(Name),
                                          'value':str(Name)} for Name in df['Name'].unique()],
                                value="L. Messi"),
                     
                     dcc.Dropdown(id="drop-down5",style={'width':'300px',"display":"inline-block"},
                                options=[{'label':str(Name),
                                          'value':str(Name)} for Name in df['Name'].unique()],
                                 value="M. Depay"),
                     
                    dcc.Graph(id='graph5',style={"border": "1px solid #A1CFEE","border-radius": "5px",'width':'650px',"display":"inline-block","margin-left": "10px"}),
                     
                    dcc.Graph(id='graph6',
                              style={"border": "1px solid #A1CFEE","border-radius": "5px",'width':'650px',"display":"inline-block","margin-right": "10px"}),
                     
                    dcc.Dropdown(id="drop-down6",style={"text-align": "center",
                                                       "margin-left": "auto", "margin-right": "auto"},
                                options=[{'label':"Bar",
                                          'value':"bar"},
                                         {'label':"Pie",
                                          'value':"pie"}],
                                value="pie"),
                    dcc.Markdown("##### Goals-Assists by the Number of Appearences",style={"text-align": "center"}),
                    dcc.Graph(id='graph-with-slider', style={"border": "1px solid #A1CFEE","border-radius": "5px"}),
                    dcc.Slider(id='Appearances-slider',
                    min=df['Appearances'].min(),
                    max=df['Appearances'].max(),
                    value=df['Appearances'].max(),
                    marks={str(Appearances): str(Appearances) for Appearances in df["Appearances"].unique()},
        step=None)])
                     
               
    

@app.callback(
Output("graph-with-dropdown", "figure"),
Input("drop-down", "value"))

def update_players_stats(inp):
    
    df_figure = df.sort_values(by=inp, ascending=False)[:10]
    bar_fig_major_cat = px.bar(data_frame=df_figure,
                               x=inp, y="Name", 
                               color="Nationality", 
                               title="Players Stats", 
                               template="seaborn",hover_name="Team Name")
    return bar_fig_major_cat
@app.callback(
    Output("graph-with-dropdown2", "figure"),
    Input("drop-down2","value"),
    Input("drop-down3","value"))

def update_team_league_stats(inp2, inp3):
    df_figure = df.sort_values(by=inp3,ascending=False)[:10]
    bar_fig_major_cat_2 = px.bar(data_frame=df_figure,
                                 x=inp2,
                                 y=inp3,
                                 color="Name",
                                 template="seaborn",
                                 title="Team-League Stats")
    
    return bar_fig_major_cat_2
@app.callback(
    Output("graph-with-dropdown3", "figure"),
    Output("graph-with-dropdown4", "figure"),
    Input("drop-down4","value"),
    Input("drop-down5","value"))

def update_comparison_graph(inp4, inp5):
    
    df_plot1 = pd.DataFrame(
                dict(r=df[df["Name"]==inp4] \
                     [['Goals',"Assists",'Key Passes','Dribbles', 'Shots On Target']].values.flatten().tolist(),
                    theta=['Goals',"Assists",'Key Passes','Dribbles', 'Shots On Target']))
    
    bar_fig_major_cat_3 = px.line_polar(df_plot1, 
                                        r='r', 
                                        theta='theta', 
                                        line_close=True,
                                        title=inp4,
                                        color_discrete_sequence=px.colors.sequential.Plasma_r)
  
    df_plot2 = pd.DataFrame(
                dict(r=df[df["Name"]==inp5] \
                     [['Goals',"Assists",'Key Passes','Dribbles', 'Shots On Target']].values.flatten().tolist(),
                theta=['Goals',"Assists",'Key Passes', 'Dribbles', 'Shots On Target']))
    
    bar_fig_major_cat_4 = px.line_polar(df_plot2, 
                                        r='r', 
                                        theta='theta', 
                                        line_close=True,
                                        title=inp5,
                                        color_discrete_sequence=px.colors.sequential.Plasma_r)
    
    return bar_fig_major_cat_3,bar_fig_major_cat_4

@app.callback(
Output("graph5", "figure"),
Output("graph6", "figure"),
Input("drop-down6","value")    
)
def update_pie_to_bar(inp6):
    if (inp6==None) or (inp6=="pie"):
        bar_fig_major_cat_6=px.pie(df_effective,names="Name",values="Goals/Min",color="Goals",title="Most Effective Players, Goals/Min",
                                                        template="ggplot2")
        bar_fig_major_cat_5=px.pie(data_frame=df.sort_values(by="Minutes Played",ascending=False)[:10],names="Name",values="Minutes Played",title="Top Players By Minutes Played",color="Position")
    else:
        bar_fig_major_cat_6=px.bar(df_effective,x="Name",y="Goals/Min",color="Goals",title="Most Effective Players, Goals/Min",
                                                        template="ggplot2")
        bar_fig_major_cat_5=px.bar(data_frame=df.sort_values(by="Minutes Played",ascending=False)[:10],x="Name",y="Minutes Played",title="Top Players By Minutes Played",color="Position")
    return bar_fig_major_cat_5,bar_fig_major_cat_6
@app.callback(
Output("graph-with-slider", "figure"),
Input('Appearances-slider','value')
)
def update_Appearances_slider(inp7):
    filtered_df=df[(df.Appearances==inp7)]
    fig1=px.scatter(filtered_df,x=filtered_df["Goals"],y=filtered_df["Assists"],size="Dribbles",color="Name",hover_name="Team Name")
    return fig1
    
    
app.run_server()
```
